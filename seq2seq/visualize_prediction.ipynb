```python
from ModelAndData import ModelDealer
%matplotlib inline
```

```python
M = ModelDealer( downcast = 500, base = 'dataset_mu10_')
```

```python
M.make_model()
```

```python
weights = M.best_model()
```

```python
data, target, truth = M.read_dataset(M.list_dataset()[10], andtruth=True)
```

```python
prediction = M.model.predict(data)
```

# show an event

```python
M.visualize( 10, data, target, prediction, truth, showTruth=False )
```

# show a series of events

```python
import random
select = range(data.shape[0])
random.shuffle( select )
```

```python
for i in select[:10]:
    M.visualize( i, data, target, prediction, truth, showTruth=False )
```

# show the evolution of an event accross training

```python
showEvent = 13
for w in weights:
    M.model.load_weights( w )
    sub_pred = M.model.predict( data )
    M.visualize( showEvent, data, target, sub_pred, truth,
                showTruth=False,
                extraTitle=w)
```
